## Result of Match

```python
def result_of_match(row, team_id):
    if team_id == row['home_club_id']:
        if row['home_club_goals'] > row['away_club_goals']:
            return 3
        elif row['home_club_goals'] < row['away_club_goals']:
            return 0
        else:
            return 1
    if team_id == row['away_club_id']:
        if row['away_club_goals'] > row['home_club_goals']:
            return 3
        elif row['away_club_goals'] < row['home_club_goals']:
            return 0
        else:
            return 1
```

## Home and Away Form during specific seasons

```python
def home_away_form(team_id, season):
    # Get games for a specific season
    games_season = games[games['season'] == season]
    home_form = games_season[games_season['home_club_id'] == team_id]
    away_form = games_season[games_season['away_club_id'] == team_id]
    
    if (len(home_form) + len(away_form)) < 14:
        return {'Home': [0, 0], 'Away': [0, 0]}
    
    # Keep only relevant columns
    home_form = home_form[['game_id', 'competition_id', 'date', 'home_club_id', 'away_club_id', 'home_club_goals', 'away_club_goals']]
    away_form = away_form[['game_id', 'competition_id', 'date', 'home_club_id', 'away_club_id', 'home_club_goals', 'away_club_goals']]

    # Win and Draw Percentage for Home
    home_form['result'] = home_form.apply(lambda row: result_of_match(row, team_id), axis = 1)
    win_draw_loss = home_form['result'].value_counts()
    win_draw_loss = win_draw_loss.reindex([3, 1, 0], fill_value = 0)
    home_win_percentage = (win_draw_loss[3] / win_draw_loss.sum()) * 100
    home_draw_percentage = (win_draw_loss[1] / win_draw_loss.sum()) * 100
    
    # Win and Draw Percentage for Away
    away_form['result'] = away_form.apply(lambda row: result_of_match(row, team_id), axis = 1)
    win_draw_loss = away_form['result'].value_counts()
    win_draw_loss = win_draw_loss.reindex([3, 1, 0], fill_value = 0)
    away_win_percentage = (win_draw_loss[3] / win_draw_loss.sum()) * 100
    away_draw_percentage = (win_draw_loss[1] / win_draw_loss.sum()) * 100

    # Return dictionary
    results = {'Home': [home_win_percentage, home_draw_percentage],
              'Away': [away_win_percentage, away_draw_percentage]}
    
    return results
```

## Last X Home and Away Match Stats

```python
# This can be configurable
def last_x_home_and_away(home_team_id, away_team_id):
    x_matches = 15

    home_matches = games[games['home_club_id'] == home_team_id]
    away_matches = games[games['away_club_id'] == away_team_id]

    # Home and Away Matches
    home_matches['date'] = pd.to_datetime(home_matches['date'])
    away_matches['date'] = pd.to_datetime(away_matches['date'])

    # Sort by most recent games
    home_matches = home_matches.sort_values(by = 'date', ascending = False)
    away_matches = away_matches.sort_values(by = 'date', ascending = False)

    # Last X Home Matches and X Away Matches
    last_x_home = home_matches.head(x_matches)
    last_x_away = away_matches.head(x_matches)

    # Results of the matches
    last_x_home['result'] = last_x_home.apply(lambda row: result_of_match(row, home_team_id), axis = 1)
    last_x_away['result'] = last_x_away.apply(lambda row: result_of_match(row, away_team_id), axis = 1)

    # Avg. Home and Away Goals
    avg_home_goals = last_x_home['home_club_goals'].sum() / len(last_x_home)
    avg_away_goals = last_x_away['away_club_goals'].sum() / len(last_x_away)

    # Avg. Win/Draw Rate at Home
    win_draw_loss = last_x_home['result'].value_counts(normalize = True)
    win_draw_loss = win_draw_loss.reindex([3, 1, 0], fill_value = 0)
    avg_home_win = win_draw_loss[3] * 100
    avg_home_draw = win_draw_loss[1] * 100
    avg_home_loss = win_draw_loss[0] * 100

    # Avg. Win/Draw Rate at Away
    win_draw_loss = last_x_away['result'].value_counts(normalize = True)
    win_draw_loss = win_draw_loss.reindex([3, 1, 0], fill_value = 0)
    avg_away_win = win_draw_loss[3] * 100
    avg_away_draw = win_draw_loss[1] * 100
    avg_away_loss = win_draw_loss[0] * 100

    # Stats
    results = {'Home': [avg_home_win, avg_home_draw, avg_home_goals], 
              'Away': [avg_away_win, avg_away_draw, avg_away_goals]}
    
    return results
```

## Referee Bias?

```python
def referee_officiating(home_team_id, away_team_id, referee):
    # Get all matches for home team
    home_team_matches = games[games['home_club_id'] == home_team_id]
    away_team_matches = games[games['away_club_id'] == away_team_id]

    # Result of Matches
    all_matches_home = home_team_matches[['game_id', 'date', 'home_club_id', 'away_club_id', 'home_club_goals', 'away_club_goals', 'referee']]
    all_matches_home['result'] = all_matches_home.apply(lambda row: result_of_match(row, home_team_id), axis = 1)
    all_matches_away = away_team_matches[['game_id', 'date', 'home_club_id', 'away_club_id', 'home_club_goals', 'away_club_goals', 'referee']]
    all_matches_away['result'] = all_matches_away.apply(lambda row: result_of_match(row, away_team_id), axis = 1)
    
    # Referee game stats for home and away
    referee_stats_home = all_matches_home.groupby('referee')['result'].value_counts().unstack(fill_value = 0)
    referee_stats_away = all_matches_away.groupby('referee')['result'].value_counts().unstack(fill_value = 0)

    # Ensure all necessary columns are present
    expected_columns = [0, 1, 3, 'total_matches', 'avg_win']
    referee_stats_home = referee_stats_home.reindex(columns = expected_columns, fill_value = 0)
    referee_stats_away = referee_stats_away.reindex(columns = expected_columns, fill_value = 0)
    
    # Referee games for home team
    referee_stats_home['total_matches'] = referee_stats_home.sum(axis = 1)
    referee_stats_home['avg_win'] = (referee_stats_home[3] / referee_stats_home['total_matches']) * 100
    referee_stats_home = referee_stats_home[referee_stats_home.index == referee]
    referee_stats_home = referee_stats_home[[3, 1, 0, 'total_matches', 'avg_win']]
    
    # Referee games for away team
    referee_stats_away['total_matches'] = referee_stats_away.sum(axis = 1)
    referee_stats_away['avg_win'] = (referee_stats_away[3] / referee_stats_away['total_matches']) * 100
    referee_stats_away = referee_stats_away[referee_stats_away.index == referee]
    referee_stats_away = referee_stats_away[[3, 1, 0, 'total_matches', 'avg_win']]
    
    # Extract home and away referee avg. wins
    avg_home = referee_stats_home.at[referee, 'avg_win'] if referee in referee_stats_home.index else 0
    avg_away = referee_stats_away.at[referee, 'avg_win'] if referee in referee_stats_away.index else 0

    return avg_home, avg_away
```

```python
# import required libraries

from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support as score, confusion_matrix, roc_auc_score, classification_report, log_loss, accuracy_score

from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

# Models to predict on
names = ["Nearest Neighbors", "Logistic Regression","Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest", "Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]

kernel = 1.0 * RBF(1.0, length_scale_bounds=(1e-2, 1e6))  # Increase upper bound

classifiers = [
    KNeighborsClassifier(3),
    LogisticRegression(),
    SVC(kernel = "linear", C=0.025, probability = True),
    SVC(gamma = 2, C = 1, probability = True),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth = 5),
    RandomForestClassifier(max_depth = 5, n_estimators = 10, max_features = 1),
    MLPClassifier(alpha=1, max_iter = 1000),
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()]
```

# Initial Thoughts

The models used for predicting football match outcomes in a multi-class framework exhibit a distinct tendency to favor home team victories, while largely overlooking the possibility of draws or away team wins. To address this issue, one effective approach is the implementation of SMOTE (Synthetic Minority Over-sampling Technique). This technique can be utilized to create synthetic examples of less frequent outcomes, such as matches ending in a draw or with an away team victory, thereby balancing the class distribution in the dataset. This balanced approach aims to reduce the models' bias towards home team wins and improve their predictive performance across all possible match outcomes.
